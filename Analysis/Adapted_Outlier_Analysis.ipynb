```python

import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
fighter_stats = pd.read_csv('../Data Cleaning/cleaned data/cleaned_ufc_fight_stat_data.csv')
```

# Outlier Analysis

## Takedowns
This section analyzes outliers in the number of successful takedowns.

```python

# Focus on the "takedown_succ" column for outlier analysis
takedowns = fighter_stats['takedown_succ']

# Calculate IQR to detect outliers
Q1_takedown = takedowns.quantile(0.25)
Q3_takedown = takedowns.quantile(0.75)
IQR_takedown = Q3_takedown - Q1_takedown

# Define lower and upper bounds for outliers
lower_bound_takedown = Q1_takedown - 1.5 * IQR_takedown
upper_bound_takedown = Q3_takedown + 1.5 * IQR_takedown

# Identify outliers
takedown_outliers = fighter_stats[(takedowns < lower_bound_takedown) | (takedowns > upper_bound_takedown)]

# Plot distribution and outliers for takedowns
plt.figure(figsize=(10, 6))
plt.boxplot(takedowns.dropna(), vert=False, patch_artist=True, showmeans=True)
plt.title("Boxplot of Successful Takedowns with Outliers")
plt.xlabel("Number of Successful Takedowns")
plt.show()

# Histogram for distribution
plt.figure(figsize=(10, 6))
plt.hist(takedowns, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(lower_bound_takedown, color='red', linestyle='--', label='Lower Bound (Outlier)')
plt.axvline(upper_bound_takedown, color='blue', linestyle='--', label='Upper Bound (Outlier)')
plt.title("Histogram of Successful Takedowns")
plt.xlabel("Number of Successful Takedowns")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Display key details about outliers
takedown_outliers_summary = takedown_outliers[['fight_stat_id', 'fighter_id', 'takedown_succ']]
takedown_outliers_summary
```

## Knockdowns
This section analyzes outliers in the number of knockdowns.

```python

# Focus on the "knockdowns" column for outlier analysis
knockdowns = fighter_stats['knockdowns']

# Calculate IQR to detect outliers
Q1_knock = knockdowns.quantile(0.25)
Q3_knock = knockdowns.quantile(0.75)
IQR_knock = Q3_knock - Q1_knock

# Define lower and upper bounds for outliers
lower_bound_knock = Q1_knock - 1.5 * IQR_knock
upper_bound_knock = Q3_knock + 1.5 * IQR_knock

# Identify outliers
knockdown_outliers = fighter_stats[(knockdowns < lower_bound_knock) | (knockdowns > upper_bound_knock)]

# Plot distribution and outliers for knockdowns
plt.figure(figsize=(10, 6))
plt.boxplot(knockdowns.dropna(), vert=False, patch_artist=True, showmeans=True)
plt.title("Boxplot of Knockdowns with Outliers")
plt.xlabel("Number of Knockdowns")
plt.show()

# Histogram for distribution
plt.figure(figsize=(10, 6))
plt.hist(knockdowns, bins=30, edgecolor='k', alpha=0.7)
plt.axvline(lower_bound_knock, color='red', linestyle='--', label='Lower Bound (Outlier)')
plt.axvline(upper_bound_knock, color='blue', linestyle='--', label='Upper Bound (Outlier)')
plt.title("Histogram of Knockdowns")
plt.xlabel("Number of Knockdowns")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Display key details about outliers
knockdown_outliers_summary = knockdown_outliers[['fight_stat_id', 'fighter_id', 'knockdowns']]
knockdown_outliers_summary
```
